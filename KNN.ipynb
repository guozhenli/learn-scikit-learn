# K Nearest Neighbor

## Import packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

## Get (fake) data
Generate fake dataset with `make_classification`

```python
X, y = make_classification(n_samples=200, n_features=2, 
                           n_informative=2, n_redundant=0, 
                           flip_y=0, random_state=666)
print(X.shape, y.shape)
```

Split data into `train` and `test` sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Plot the training set:

```python
plt.figure()
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, s=50)
plt.title('Training data set')
plt.show()
```

```python
plt.figure()
plt.scatter(X_test[:,0], X_test[:,1], c=y_test, s=50)
plt.title('Test data set')
plt.show()
```

## Train KNN model

The `KNeighborsClassifier` class comes from `sklearn.neighbors` module. 
Create such an object, and fit it with our train data.

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

See how the model does on the test set

```python
knn.score(X_test, y_test)
```
